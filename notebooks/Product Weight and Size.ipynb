### Product Weight, Size

Analyzing wherether or not `product_weight`, `product_size` have impact on delivery time.

```python
import pandas as pd

from utils import read_product_delivery
```

```python
dataset = read_product_delivery()
dataset.head()
```

```python
print(dataset.columns)
```

```python
cols = ['quantity', 'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm', 'freight_value']
df = dataset[cols].copy()
```

```python
df['total_days'] = dataset['total_delivery_time'].dt.days
```

```python
df.corr()
```

#### Finding 
In statistics, a variable can be a powerful predictor even if it feels like "common sense." In fact, since product_weight has a weak correlation but freight_value has a significant one, your data is telling you something very important: Distance or Carrier Type is more important than the physical weight of the item.
