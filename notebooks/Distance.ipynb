### Delivery Time Geolocation

According to previous analysis, the average delivery time differs base on states.

However, there are extremely late deliveries regardless of states. 

In this analysis, we'll try to find fault beyond pinning it to states. I'll try to find which municiplities suffer more than other or not.
If some municiplities are worse than the others, I'll try to find the behaviour in it.

    

```python
import pandas as pd
import h3

from utils import read_customer_seller
```

```python
dataset = read_customer_seller()
dataset.head()
```

```python
df = dataset[['cus_state', 'seller_state']].copy()
df['total_days'] = (dataset['order_delivered_customer_date'] - dataset['order_purchase_timestamp']).dt.days
```

```python

# Calculate Distance
df['dist'] = dataset.apply(lambda row: h3.great_circle_distance((row['cus_lat'], row['cus_lng']), (row['seller_lat'], row['seller_lng'])), axis=1)

df.head()
```

```python
df[['total_days', 'dist']].corr()
```
