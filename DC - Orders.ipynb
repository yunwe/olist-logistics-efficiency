```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
orders = pd.read_csv("DataSet/olist_orders_dataset.csv" 
                     , parse_dates=[
                         'order_purchase_timestamp', 
                         'order_approved_at',
                         'order_delivered_carrier_date',
                         'order_delivered_customer_date',
                         'order_estimated_delivery_date'
                     ])
```

```python
orders.head()
```

```python
orders['order_status'].value_counts()
```

```python
orders['total_delivery_time'] = orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']
orders['estimated_delivery_time'] = orders['order_estimated_delivery_date'] - orders['order_purchase_timestamp']
orders['wait_approve_time'] = orders['order_approved_at'] - orders['order_purchase_timestamp']
orders['seller_to_logistic_time'] = orders['order_delivered_carrier_date'] - orders['order_approved_at']
orders['logistic_to_customer_time'] = orders['order_delivered_customer_date'] - orders['order_delivered_carrier_date']

orders.head()
```

```python
total_hours = orders['total_delivery_time'].dt.total_seconds() / 3600.0
total_hours.head()
```

```python
def save_file(df, filename):
    now = datetime.now()

    # 2. Format it (YYYY-MM-DD_HH-MM-SS)
    # %Y = Year, %m = Month, %d = Day, %H = Hour, %M = Minute, %S = Second
    timestamp = now.strftime("%Y-%m-%d_%H-%M-%S")

    fname = f"DataSet/{filename}_{timestamp}.csv"
    df.to_csv(fname, index=False)
```

```python
save_file(orders, "orders")
```
